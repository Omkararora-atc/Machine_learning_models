### Spliting the dataset into features and target variable

```python
x=df.drop('quality',axis=1)
y=df['quality']
```

```python
x
```

### Checking for outliers

```python
fig,ax=plt.subplots( figsize=(20, 10))
sns.boxplot(data=x, ax=ax)
plt.xticks(rotation=45)
plt.title('Boxplot of Features')
plt.savefig('boxplot.png')
```

```python
x.columns
```

```python
col=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
mask = pd.Series(True, index=x.index)

for i in col:
    q1 = x[i].quantile(0.25)
    q3 = x[i].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    mask = (x[i] > lower_bound) & (x[i] < upper_bound)  # Combine masks
x_outlier_detection = x[mask]
y_outlier_detection = y[mask]
```

```python
x_outlier_detection.shape
```

```python
fig,ax=plt.subplots( figsize=(20, 10))
sns.boxplot(data=x_outlier_detection, ax=ax)
plt.xticks(rotation=45)
plt.title('Boxplot of Features')
plt.show()
```

#### Treating outlier in the total sulfur dioxide column

```python
# Calculate the 95th percentile value for 'total sulfur dioxide'
percentile_95 = x_outlier_detection['total sulfur dioxide'].quantile(0.95)

# Cap the values above the 95th percentile
x_outlier_detection['total sulfur dioxide'] = np.where(
    x_outlier_detection['total sulfur dioxide'] > percentile_95,
    percentile_95,
    x_outlier_detection['total sulfur dioxide']
)
```

```python
fig,ax=plt.subplots( figsize=(20, 10))
sns.boxplot(data=x_outlier_detection, ax=ax)
plt.xticks(rotation=45)
plt.title('Boxplot of Features')
plt.show()
```

### spliting the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_outlier_detection, y_outlier_detection, test_size=0.2, random_state=42)
```

```python
x_test.shape
```

```python
x_train.shape
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
fig,ax=plt.subplots( figsize=(20, 10))
sns.boxplot(data=x_train, ax=ax)
plt.xticks(rotation=45)
plt.title('Boxplot of Features')
plt.show()
```

#### Model Building

```python
from sklearn.tree import DecisionTreeClassifier

model= DecisionTreeClassifier(class_weight='balanced', random_state=42)

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(20, 10))
tree.plot_tree(model,filled=True)
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
