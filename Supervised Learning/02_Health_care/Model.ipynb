### Filling Null Values

```python
df['Insulin'] = df['Insulin'].replace(0, df['Insulin'].median())
df['Pregnancies'] = df['Pregnancies'].replace(0, df['Pregnancies'].median())
df['Glucose'] = df['Glucose'].replace(0, df['Glucose'].mean())
df['BloodPressure'] = df['BloodPressure'].replace(0, df['BloodPressure'].mean())
df['SkinThickness'] = df['SkinThickness'].replace(0, df['SkinThickness'].median())
df['BMI'] = df['BMI'].replace(0, df['BMI'].mean())
df['DiabetesPedigreeFunction'] = df['DiabetesPedigreeFunction'].replace(0, df['DiabetesPedigreeFunction'].median())
df['Age'] = df['Age'].replace(0, df['Age'].median())
```

### Checking for outliers

```python
x=df.drop('Outcome', axis=1)
y=df['Outcome']
```

```python
fig,ax=plt.subplots( figsize=(20, 10))
sns.boxplot(data=x, ax=ax)
plt.xticks(rotation=45)
plt.title('Boxplot of Features')
plt.savefig('boxplot.png')
```

### Treatment of Outliers

```python

col = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
mask = pd.Series(True, index=x.index)

for i in col:
    q1 = x[i].quantile(0.25)
    q3 = x[i].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    mask= (x[i] >= lower_bound) & (x[i] <= upper_bound)

# Apply the combined mask
x_outlier_detection = x[mask]
y_outlier_detection = y[mask]
```

```python
x_outlier_detection.shape
```

```python
y_outlier_detection.shape
```

### standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x_outlier_detection)
```

```python
fig,ax=plt.subplots( figsize=(20, 10))
sns.boxplot(data=X_scaled, ax=ax)
plt.xticks(rotation=45)
plt.title('Boxplot of Features')
plt.savefig('boxplot.png')
```

### Outlier Treatment

```python

X_scaled = pd.DataFrame(X_scaled, columns=x_outlier_detection.columns)
X_scaled.reset_index(drop=True, inplace=True)
y_outlier_detection = y_outlier_detection.reset_index(drop=True)
```

```python
q = X_scaled['Insulin'].quantile(.95)
mask = X_scaled['Insulin'] < q
dataNew = X_scaled[mask]
y_outlier_detection = y_outlier_detection[mask]
```

```python
dataNew.shape
```

```python
y_outlier_detection.shape
```

```python
fig,ax=plt.subplots( figsize=(20, 10))
sns.boxplot(data=dataNew, ax=ax)
plt.xticks(rotation=45)
plt.title('Boxplot of Features')
plt.savefig('boxplot.png')
```

### Splitting the Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataNew, y_outlier_detection, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train.value_counts()
```

### Observations

* The dataset is imbalanced, with more instances of class 0 (non-diabetic) than class 1 (diabetic).
* we use the Synthetic Minority Over-sampling Technique (SMOTE) to balance the dataset.


```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_bal,y_train_bal = smote.fit_resample(X_train,y_train)
```

```python
print(pd.Series(y_train_bal).value_counts())
```

### Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
classifier.fit(x_train_bal, y_train_bal)
```

```python
y_ped= classifier.predict(X_test)
```

```python
y_ped
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_ped)
```

```python
accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_ped))
```
